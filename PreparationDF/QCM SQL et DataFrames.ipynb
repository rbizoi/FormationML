```python
donnees['SumTrim'] = donnees.groupby(['annee','trimestre']).transform(sum)['port_sum']
donnees['SumAnnee'] = donnees.groupby(['annee']).transform(sum)['port_sum']
donnees.set_index(['annee','trimestre','mois'],inplace=True)
```
