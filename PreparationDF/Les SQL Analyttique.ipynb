La lecture d'une table et le chargement dans un DataFrame

```python
print("connecting with engine " + str(engine))
connection = engine.connect()
query = """
select grouping(code_client) * 4 +
       grouping(annee) * 2 + 
       grouping(trimestre) gi3,
       grouping_id(code_client, annee, trimestre) gi,
       code_client, annee, trimestre, sum(port) as port
from commandes
group by code_client,rollup ( annee, trimestre)
"""
df = pd.read_sql_query(query, connection) #df.to_sql('nom', con=engine,if_exists='replace')
df
```
